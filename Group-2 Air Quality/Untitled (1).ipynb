```python
# importing libaries
import pandas as pd
import altair as alt
```

```python
# Read the filtered_fire_data.csv file into a Pandas DataFrame
df = pd.read_csv('chestersquareair.csv')
```

```python
#summing all air quaility 44.1
df['Sum'] = df['Air quality improved (lb/year)'].sum()
```

```python
# new data frame for one tree
# 2950114, Northern Red Oak
source1 = pd.DataFrame({"Tree": ['Northern Red Oak', 'Total Trees in Chester Square Park'], "Air quality improved (lb/year)": [3.2, 40.9]})
```

```python
# Pie chart for 2950114, Northern Red Oak
base = alt.Chart(source1).encode(
    alt.Theta("Air quality improved (lb/year)").stack(True),
    alt.Color("Tree", scale=alt.Scale(range=["lightgreen", "darkgreen"])).legend(None)
).properties(
    width=400,
    height=400,
    title={
      "text": ["Pounds of Pollution Removed by Trees in Chester Square Park"], 
      "subtitle": ["(Air quality improved lb/year)"]
    }
)
pie = base.mark_arc(outerRadius=120)
text = base.mark_text(radius=80, size=15).encode(text="Air quality improved (lb/year):N", color=alt.value("black"))
text2 = base.mark_text(radius=140, size=20).encode(text="Tree:N")

pie + text + text2
```
